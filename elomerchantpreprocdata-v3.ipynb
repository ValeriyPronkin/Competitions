```python
#Количество магазинов где были покупки по карте
def agg_count_sale(data):
   count_sale = data.value_counts().count()
   return count_sale

#Максимальное количество покупок в одном магазине
def agg_max_sale(data):
    max_sale = data.value_counts().max()
    return max_sale
##Минимальное количество покупок в одном магазине
def agg_min_sale(data):
    min_sale = data.value_counts().min()
    return min_sale
#Общее количество покупок по карте во всех магазинах
def agg_sum_sale(data):
    sum_sale = data.value_counts().sum()
    return sum_sale
#среднее количество покупок в одном магазине
def agg_mean_sale(data):
    mean_sale = data.value_counts().mean()
    return mean_sale
```
