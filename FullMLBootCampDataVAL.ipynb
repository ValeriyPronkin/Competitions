```python
def df_feat_prep(df_fe, aggfunc=np.sum, values='REVENUE'):
    i=1
    dfe_1 = df_fe.pivot_table(values=values, index=["SK_ID"],
                            columns=["SNAP_DATE"],
                              aggfunc=aggfunc,
                              dropna=False)
    dfe_1.rename(columns={'01.01.02':'8_#'+str(i), '01.02.02':'9_#'+str(i), '01.03.02':'10_#'+str(i)
                          , '01.04.02':'11_#'+str(i), '01.05.02':'12_#'+str(i), '01.06.01':'1_#'+str(i)
                          , '01.07.01':'2_#'+str(i), '01.08.01':'3_#'+str(i), '01.09.01':'4_#'+str(i)
                          , '01.10.01':'5_#'+str(i), '01.11.01':'6_#'+str(i), '01.12.01':'7_#'+str(i)},
                 inplace=True)
    dfe1=dfe_1[['1_#'+str(i), '2_#'+str(i), '3_#'+str(i), '4_#'+str(i), '5_#'+str(i), '6_#'+str(i)
                , '7_#'+str(i), '8_#'+str(i), '9_#'+str(i), '10_#'+str(i), '11_#'+str(i), '12_#'+str(i)]]

    dfe1=dfe1.apply(lambda x: x.fillna(x.map(meg_preproc).min()),axis=1)
    return dfe1
```

```python
def df_cons_prep(df_bs, on='SUM_MINUTES', aggfunc=np.sum,):
    df_bs['SUM_MINUTES'] = df_bs['SUM_MINUTES'].map(meg_preproc).astype(np.float64)
    df_bs['SUM_DATA_MB'] = df_bs['SUM_DATA_MB'].map(meg_preproc).astype(np.float64)
    df_bs['SUM_DATA_MIN'] = df_bs['SUM_DATA_MIN'].map(meg_preproc).astype(np.float64)
    i=on
    dbs_1 = df_bs.pivot_table(values=i, index=["SK_ID"],
                              columns=["MON"], 
                              aggfunc=aggfunc,
                              dropna=False)
    col=[]
    for p in dbs_1.columns:
        col.append(str(p)+'_'+str(i))
    dbs_1.columns = col
    dbs1=dbs_1.apply(lambda x: x.fillna(x.min()),axis=1)
    return dbs1
```

## Trying new agg func

```python
## moving average diffirence
def AGG_DIFF_MA(x, gamma=0.5):
    res = 0
    prev_elem = x.iloc[0]
    for elem in x:
        res = res * (1 - gamma) + (elem - prev_elem) * gamma
        prev_elem = elem
    return res
```

```python
def AGG_DIFF(x):
    res = 0
    prev_elem = x.iloc[0]
    for elem in x:
        res += elem - prev_elem
        prev_elem = elem
    return res
```

```python
def AGG_MA(x, gamma=0.8):
    res = x.iloc[0]
    for elem in x:
        res = res * (1 - gamma) + elem * gamma
    return res 
```

```python
def AGG_DIFF_ABS(x):
    res = 0
    prev_elem = x.iloc[0]
    for elem in x:
        res += abs(elem - prev_elem)
        prev_elem = elem
    return res
```
